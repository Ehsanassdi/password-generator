```python
# -----------------------------------------
# Password Generator Program
# Author: Ehsan Assadi
# Description:
# This program generates a random password based on user input.
# The user specifies the desired password length (4â€“128),
# and the program creates a secure password using lowercase,
# uppercase, digits, and special characters.
# The user can choose to generate another password or quit.
# ----------------------------------


if __name__ == "__main__":
    # Import required libraries
    import random
    import string
    
    # Start the main program loop
    while True:
            # Step 1: Ask for password length or allow quitting
            password_num = input("Please enter the password length(4 - 128: ")
            if password_num.lower() == "q" or password_num.lower() == "quit":
                print('goodbye')
                break

            # Step 2: Validate that the input is a number
            elif  not password_num.isdigit():
                    print("Please enter a valid number")
                    continue
                    
            # Step 3: Convert input to integer and validate range
            else:    
                num = int(password_num)
                password = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation
                if num < 4 or num > 128:
                    print("The number is too small or too large")
                    continue

                # Step 4: Generate password using random choices
                generated_password = ''.join(random.choices(password, k=num))
                print(f"Your generated password is: {generated_password}")

                # Step 5: Ask if the user wants to generate another one
                choice = input("Do you want to continue? Y/N")
                if choice.lower() != "y":
                    print("Exiting the program.")
                    break
```
